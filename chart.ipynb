# Charts of network monitor
Use pandas to draw the CSV data from network monitor.

```python
import pandas as pd
import matplotlib.pyplot as plt
def plot_csv(title, csvfile, cumsum=False, saveimg=None):
    csv = pd.read_csv(csvfile)
    df = csv.set_index('Time')
    df.index = df.index.map(lambda x: x / 1000.0)
    if cumsum:
        df = df.cumsum()
    ax = df.plot(figsize=(10, 7))
    ax.xaxis.grid(True, which="major")
    ax.yaxis.grid(True, which="major")
    ax.set_xlabel("Time(ms)")
    ax.set_ylabel("Value")
    ax.set_title(title)
    plt.show()
    if saveimg:
        fig = ax.get_figure()
        fig.savefig(saveimg) 
```

```python
plot_csv("Received bytes of NICs", "data/netmon.csv")
```

```python
plot_csv("Accumulated received bytes of NICs", "data/netmon.csv", cumsum=True, saveimg="image/netmon.png")
```
